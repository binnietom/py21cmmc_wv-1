```python
%load_ext autoreload
%autoreload 2

from py21cmmc_wv.likelihood import LikelihoodWaveletsMorlet
from py21cmmc.mcmc.mcmc import run_mcmc
from py21cmmc.mcmc.core import CoreLightConeModule
import os
import sys

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
model_name = "runthrough_test"

core = CoreLightConeModule(
    redshift=7.0,  #probably make it 6
    max_redshift=9.0, #probably make it 12
    user_params=dict(
        HII_DIM=25,
        BOX_LEN=50.0
    ),
    regenerate=False,
)

likelihood = LikelihoodWaveletsMorlet(
    datafile = f"data/{model_name}.npz",
    simulate=True
)

chain = run_mcmc(
    core, likelihood,
    datadir='data',
    model_name=model_name,
    params = dict(
        HII_EFF_FACTOR=[30.0, 10., 50.0, 3.0],
        ION_Tvir_MIN = [4.7, 2, 8, 0.1]
    ),
    walkersRatio=8,
    burninIterations=0,
    sampleIterations=100,
    threadCount=2,
    continue_sampling=False
)
```

# Show Covariance

```python
cov = likelihood.compute_covariance(
    likelihood.data[0]['wavelets'], 
    likelihood.data[0]['kpar'], 
    likelihood.data[0]['centres'])
```

```python
plt.imshow(cov[0].T, origin='lower')
plt.colorbar()
```
